### **Source de données 1 (df_perf) : Performance des joueurs**

```python
df_perf = pd.read_csv("foot_perf.csv")
df_perf
```

### **Source de données 2 (df_mv) : Valeur marchande des joueurs**

```python
df_mv = pd.read_csv("only_mv.csv")
df_mv = df_mv.drop(columns = df_mv.columns[0], axis = 1)
df_mv = df_mv.drop(["Nation", "Pos"], axis = 1)
df_mv
```

### Focus sur les données complété côté Market value rattaché à la performance du joueur

```python
merged_df = pd.merge(df_perf, df_mv, on='Player', how='left')
merged_df
```

```python
df_filtered = merged_df[merged_df["Market value"].notnull()]
df_filtered
```

On remarque que après jointure des deux en commun. On a réduit massivement les données à disposition. Nous travaillerons de ce fait sur le remplissage des valeurs manquantes

### **Nettoyage et analyse du jeu de données "merged_df"**

```python
merged_df.isnull().sum().sum()
```

```python
merged_df.columns[merged_df.isnull().any()].tolist()
```

```python
merged_df['Pos'].unique()
```

```python
merged_df['Nation'].unique()
```

```python
merged_df['market_int'] = merged_df['Market value'].str.replace('m', '').str.replace("€", "")
merged_df['market_int'] = merged_df['market_int'].astype(float)
merged_df
```

```python
median_value = merged_df['market_int'].median()
merged_df['market_int'] = merged_df['market_int'].fillna(round(median_value/2,2))
merged_df
```

On remplace la valeur manquante au niveau de la valeur marchande par la moitié de la médiane de notre échantillon. Afin de ne pas biaiser les valeurs existantes

```python
merged_df = merged_df.drop_duplicates(["Player"], keep = "first")
merged_df
```

On remarque des valeurs en double dans la colonne 'nom des joueurs' ("Player"), on décide donc de garder le premier élément qui apparaît et de supprimer le second. Ceci étant expliqué par le fait qu'un joueur peut changer d'équipe en cours de saison, ce qui montre pourquoi ce joueur apparaît deux fois dans le dataset et possède des performances différentes.

## **Étude de nos données (df_football_stats)**

```python
df_football_stats = pd.DataFrame()
```

```python
df_football_stats['Player'] = merged_df['Player']
df_football_stats['Nation'] = merged_df['Nation']
df_football_stats['Pos'] = merged_df['Pos']
df_football_stats['Squad'] = merged_df['Squad']
df_football_stats['Comp'] = merged_df['Comp']
df_football_stats['Age'] = merged_df['Age']
df_football_stats['MarketValue (M €)'] = merged_df['market_int']
df_football_stats['MP'] = merged_df['MP']
df_football_stats['G/90'] = merged_df['Goals']
df_football_stats['G/Sh'] = merged_df['G/Sh']
df_football_stats['PKGoals'] = ((merged_df['ShoPK'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['PKAttempted'] = ((merged_df['PKatt'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Goals'] = ((merged_df['Goals'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Pass'] = ((merged_df['PasTotAtt'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['PassCompleted'] = ((merged_df['PasTotCmp'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['PassComp%'] = ((df_football_stats['PassCompleted'] / df_football_stats['Pass']) * 100).round(2)
df_football_stats['Assist'] = ((merged_df['Assists'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Cross'] = ((merged_df['PasCrs'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['CrossCompleted'] = ((merged_df['CrsPA'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['CrossComp%'] = ((df_football_stats['CrossCompleted'] / df_football_stats['Cross']) * 100).round(2)
df_football_stats['Tackle_Won'] = ((merged_df['TklWon'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['SucDribble'] = ((merged_df['DriSucc'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Dribble'] = ((merged_df['DriAtt'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['DribbleComp%'] = ((df_football_stats['SucDribble'] / df_football_stats['Dribble']) * 100).round(2)
df_football_stats['YCards'] = ((merged_df['CrdY'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['RCards'] = ((merged_df['CrdR'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Fls'] = ((merged_df['Fls'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['Fld'] = ((merged_df['Fld'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['OGoals'] = ((merged_df['OG'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['AerWon'] = ((merged_df['AerWon'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['AerLost'] = ((merged_df['AerLost'] * merged_df['Min']) / 90).round(0).astype(int)
df_football_stats['AerWon%'] = ((merged_df['AerWon'] / (merged_df['AerWon'] + merged_df['AerLost'])) * 100).round(2)
```

### Top 10 des meilleurs buteurs

```python
(df_football_stats[(df_football_stats.MP >= 20) & (df_football_stats.Pos != 'GK')].sort_values('Goals', ascending = False).head(10).iloc[:, [0,1,2,3,4,5,6,7,12]])
```

### Top 5 des clubs qui marquent le +

```python
team_goals = df_football_stats.groupby(['Squad']).agg(Team_Goals = ('Goals', 'sum'))
team_goals = pd.DataFrame(team_goals).reset_index()
team_goals.sort_values('Team_Goals', ascending = False).head(5)
```

### Top 10 Players with most Assists

```python
df_football_stats.sort_values('Assist', ascending = False).head(10).iloc[:, [0,1,2,3,4,5,6,16]]
```

### Top 5 des clubs qui ont le + d'assist

```python
team_assist = df_football_stats.groupby(['Squad']).agg(Team_Assist = ('Assist', 'sum'))
team_assist = pd.DataFrame(team_assist).reset_index()
team_assist.sort_values('Team_Assist', ascending = False).head(5)
```

### Top 5 des clubs qui ont reçu le plus de carton jaune

```python
team_yc = df_football_stats.groupby(['Squad']).agg(Team_YCards = ('YCards', 'sum'))
team_yc = pd.DataFrame(team_yc).reset_index()
team_yc.sort_values('Team_YCards', ascending = False).head(5)
```

### Top 10 des clubs qui ont les joueurs dotés d'une valeur marchande élevé

```python
team_mv = df_football_stats.groupby(['Squad']).agg(Team_MValue = ('MarketValue (M €)', 'sum'))
team_mv = pd.DataFrame(team_mv).reset_index()
team_mv.sort_values('Team_MValue', ascending = False).head(10)
```

### Top 5 des nations qui ont les joueurs dotés d'une valeur marchande élevé

```python
df_mv_sorted = df_football_stats.sort_values('MarketValue (M €)', ascending = False)

# group by nation and get the top 5 players by market value
df_top5 = df_mv_sorted.groupby("Nation").agg(Team_MValue = ('MarketValue (M €)', 'sum'))
interteam_mv = pd.DataFrame(df_top5).reset_index()
interteam_mv.sort_values('Team_MValue', ascending = False).head(10)
```
